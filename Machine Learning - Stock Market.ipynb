Importing the yfiance library

```python
import yfinance as yf
```

Create a Ticker object for the S&P 500 index, with the symbol "^GSPC".

```python
sp500 = yf.Ticker("^GSPC")
```

Retrieve the historical data for the S&P 500 index and store it in the sp500 variable.

```python
sp500 = sp500.history(period="max")
```

Plot a line graph of the closing prices from the sp500 data.

```python
sp500.plot.line(y="Close", use_index = True)
```

Let's clean the data by removing the 'Dividends' and 'Stock Splits' columns from the sp500 data.

```python
del sp500['Dividends']
del sp500['Stock Splits']
```

Create a new column in the sp500 data called 'Tomorrow' by shifting the 'Close' column up by one day.

```python
sp500["Tomorrow"] =sp500["Close"].shift(-1)
sp500
```

We converted the tomorrow column so that it shows a 1 if the price went up from the previous day or a 0 if it went down

```python
sp500['Target'] = (sp500['Tomorrow'] > sp500['Close']).astype(int)
sp500
```

We removed data that is older than 1990 to aviod market shifts/inaccuracy

```python
sp500 = sp500.loc['1990-01-01':].copy()
sp500
```

```python
# Import the RandomForestClassifier class from the sklearn.ensemble module.
from sklearn.ensemble import RandomForestClassifier

# Create a RandomForestClassifier model with specified parameters.
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

# were splitting the model into 2
# frist 100 rows in train
train = sp500.iloc[:-100]

# second 100 in test
test = sp500.iloc[-100:]

# Define a list of predictors.
predictors = ["Close", "Volume", "Open", "High", "Low"]

# Train the model using the training data.
model.fit(train[predictors], train['Target'])
```

Import the precision_score function from the sklearn.metrics module.
Generate predictions using the model and the testing data.

```python
from sklearn.metrics import precision_score 
preds = model.predict(test[predictors])
```

Create a Pandas DataFrame for the predictions, including the 'Target' and 'Predictions' columns.

```python
import pandas as pd
preds = pd.Series(preds, index = test.index)
```

```python
precision_score(test['Target'], preds)
```

Let's build a more robust way to test. 

```python
# Define a function called predict to train a model and generate predictions.
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index = test.index, name = 'Predictions')
    combined = pd.concat([test['Target'], preds], axis = 1)
    return combined
```

We first start off with the first ten years of data, then use that to predict the 11th year of data, then take those 11 years and use it to predict the 12th and so on.

```python
# Define a function called backtest to perform a rolling prediction on the data.
def backtest(data, model, predictors, start = 2500, step = 250):
    #list of predictions
    all_predictions = []
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

Perform a backtest using the sp500 data, model, and predictors, and store the predictions.

```python
predictions = backtest(sp500, model, predictors)
```

Add more predictors to imporve accuracy

```python
horizons = [2,5,60,250,1000]
new_predictors = []
for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500['Close'] / rolling_averages['Close']

    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna()
sp500
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

Modify the predict function to use predict_proba for probability predictions.

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    # swapped out mode.predict with model.predict_proba so it gives us the probablity of returning a 1 or 0
    preds = model.predict_proba(test[predictors])[:,1]
    # we made the thresholds bigger, so that the model is more confident before it predicts
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index = test.index, name = 'Predictions')
    combined = pd.concat([test['Target'], preds], axis = 1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
# Calculate the precision score of the adjusted predictions.
precision_score(predictions['Target'], predictions["Predictions"])
```
